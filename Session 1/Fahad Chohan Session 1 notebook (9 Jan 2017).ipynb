This code is used to convert polar coordinates to cartesian coordinates and is then used to plot a galilean spiral.

I have imported numpy and matplotlib. 

Also, the "%matplotlib inline" function is a command which keeps the data plots in  the same notebook and not a new pop-up

```python
#Importing NumPy
import numpy as np

#Importing matplotlib
import matplotlib.pyplot as plt

#Keeping matplotlib inside notebook
%matplotlib inline
```

Now we need to define a function that converts polar coordinates to cartesian coordinates

```python
def cartesian(r,theta):
    """Function that finds the cartesian valuea for a set of polar coordinates"""
    
    x = r*np.cos(theta) #calculates x values
    y = r*np.sin(theta) #calculates y values
    return x,y #returns values for x and y
```

Formula for the galilean spiral is:
$$r = \theta^2 - l$$

Now we define values for $\theta$ and $l$

Then calculate values for $r$ using the above galilean formula

$\theta$ is between $0$ and $10\pi$

$l$ is set to $0.1$

Finally we plot a graph for this range of values

```python
theta_1 = np.linspace(0,10*np.pi,1000) #creates an array for theta with 1000 values between 0 and 10pi
l_1 = 0.1 #defines a value for l

r_1 = theta_1**2 - l_1 #Galilean spiral formula
x_1,y_1 = cartesian(r_1,theta_1) #finds cartesian values for galilean spiral

plt.plot(x_1,y_1) #plots x_1 and y_1 array
plt.title('Galilean spiral') #Gives a title to plot
plt.xlabel('x') #x-axis label
plt.ylabel('y') #y-axis label
```

Now we will plot five different galilean spirals on the same graph for 5 evenly spaced values for $l$

```python
l_2 = 0 #set parameter l_2 to zero
a = 0 #this parameter will be used to label the different spirals
theta_2 = np.linspace(0,2*np.pi,1000) #Using a new range for theta, between 0 and 2pi with a 1000 values

plt.title('Galilean spirals for five different values of L') #defines graph title
plt.xlabel('x') #labels x-axis
plt.ylabel('y') #labels y-axis


#Now we will create a while loop that will be used to plot 5 galilean spirals for 5 values of l_2
while l_2 <= 10: #while loop for l_2 equal to or less than 10
    r_2 = theta_2**2 - l_2 #calculates an r value for current l_2
    x_2,y_2 = cartesian(r_2,theta_2) #uses defined cartesian function above to calculate cartesian values for r_2 and theta_2
    plt.plot(x_2,y_2, label=a) #plots graph and uses parameter a to label current spiral
    plt.legend(loc="lower left") #plots legend
    a = a + 2.5 #increases value of a by 2.5 so that label value changes and so each label is distinguished
    l_2 = l_2 + 2.5 #increases l_2 by 2.5 so that next spiral cant be plotted with a new l_2 value
```
